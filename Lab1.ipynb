```python
import datetime
import pandas as pd
```

```python
weather = pd.read_csv('last_ten_alb.csv')
weather.head()
weather['timestamp'] = pd.to_datetime(weather['timestamp'])
weather.head()
```

```python
weather.describe()
```

```python
daily_avg_tmp = weather.groupby(pd.Grouper(key='timestamp', freq='D')).mean()
daily_avg_tmp
daily_avg_tmp.plot()

newt = weather.groupby(pd.Grouper)
```
